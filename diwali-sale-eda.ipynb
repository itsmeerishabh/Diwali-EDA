```python
import seaborn as sns
sns.countplot(data = df, x = "Gender")
```

```python
sales_agegroup = pd.DataFrame(df.groupby(["Age Group"],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False))
sales_agegroup
```
